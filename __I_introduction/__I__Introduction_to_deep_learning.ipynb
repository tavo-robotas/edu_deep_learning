### Typical supervised learning workflow 

We are going to use supervised learning quite a lot in upcomming weeks. I just want to briefly outline how supervised learning works. It is a process of `regression`or `classification` so there are essentially two steps. One is `training` step , so we train the machine learning model and second step is `inference` step. But before we to the inference lets focus on training. 
<br>
Imagine we have a training data set and our task in this case is classification of flowers. We have different flowers species. And we are interested to classify what type of flower it is based on image. So we have collected multiple flower images and thats our training data set and we also have expert that provides us with the labels for those flowers like iris, sunflower and rose. Thats our `labels` that we want to predict and the sample we have collected are our `observations`.  And in traditional machine learning what we do is extract some features ussually traditional ML algorithms don't work very well on image data ,but lets keep going. For feature extraction we would need domain expert on what is good feature. Maybe the potentially good feature would be the color of the flower and maybe the hight and other aspects like number of leaves. We have to think about what could be a good feature that we can obtain from image data easily. 
<br>
So we have now `features` and `labels` then we can train an algorithm to make predictions. So after training the model a.k.a `fitting` the model or `parameterizing` , depends on what model we are talking about fyi some model are not parametric so we really don't update `weight` parameters in the model. But for now just remember that all those terms are synonymous in the context of deep learning. 
<br>

Now we have our model that was parameterized based on the training dataset. Then goal is to make predictions on new `observations` new flowers. Like imagine that you have developed a website where people can upload pictures of flowers and then your pipeline can predict what type of flower it is. You would get new images of flowers there would be feature extraction step for these flowers and it exactly the same as durring the training phase. Because our model has learned to work with this data that it has seen durring the training so we have to have the same feature extractions steps. And then it predicts new labels. This is nowadays called `inference` though it is different from statistical inference but it just happens to be called that way. But just think of it as a prediction on new data.

### Evaluating the performance of a predictive model 

We also somehwere in this process what to evaluate how good is our model before we employ it in the real world. This is relative to the inference step however use a `test set` for that. This set looks like a training set but it is independant from the training set. If we have large dataset ussually we shuffle the dataset and randomly divide it into large portions for training and smaller portion for testing. Ussualy its like apprx. 80:20 ratio it depends on the dataset size. But there is a third split `validation set` that we use for tunning the model. So we have a ratios of <code>**70:10:20**</code>. We will talk about it more later when will do our first code examples. For now consider this test set that looks very similar to the training set, we have observations and labels and now we do the inference step. But we have the `knowned labels` so we separate `observation` and `labels`, we withold them from the model because there is no training now. And then we use observations as new (we pretend that they are new). Then we this data we do feature extraction and give it to the model and the model makes the predictions. We end up with two sets of labels, one is knowned labels (growned truth data) and predicted labels set. We compare them and count the number of times how ofter our model makes a correct prediction and then we compute the `accuracy` by dividing the correct predictions with the total of observations in testing set. Similarly we can compute `error` by 1 - accuracy.
