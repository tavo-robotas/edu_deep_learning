### Data set preparation

```python
df = pd.read_csv('../data_samples/iris.data', index_col=None, header=None)
```

```python
df.columns = ['x1', 'x2', 'x3', 'x4', 'y']
```

```python
df.y = df.y.apply(lambda x: 0 if x == 'Iris-versicolor' else 1)
```

### Assign features and target

```python
xs = df[['x2', 'x3']].values
xs
```

```python
ys = df['y'].values
ys
```

```python
X = torch.tensor(xs, dtype=torch.float)
y = torch.tensor(ys, dtype=torch.float)
```

### Shuffling and train test split

```python
torch.manual_seed(123)
idx = torch.randperm(y.size[0], dtype=torch.long)
```

```python
X, y = X[idx], y[idx]
```

```python
p70 = int(idx.size(0) *.7)
```

```python
X_train, X_test = X[idx[:p70]], X[idx[p70:]]
y_train, y_test = y[idx[:p70]], y[idx[p70:]]
```

### Normalize mean zero unit variance

```python
mu , sigma  = X_train.mean(dim=0), X_train.std(dim=0)
```

```python
X_train_norm = (X_train - mu) / sigma
X_test_norm  = (X_test - mu) / sigma
```
