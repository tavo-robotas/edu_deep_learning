We need a learning rule that would be used to determine a good <code>**threshold**</code> and good <code>**weights values**</code>. So that we can solve different classification problems. ### Terminology

<code>**General (logistic regression, multilayer nets, ...):**</code>
 - Net input = weighted inputs, z
\begin{multline*}
\sum^{m}_{i=1} x_{i}w_{i}
\end{multline*}
 - Activations = activation function(net input)
\begin{multline*}
a = \delta(z)
\end{multline*}
 - Label output = threshold(activations of last layer); 
\begin{multline*}
\hat{y} = f(a)
\end{multline*}
_________________________
\begin{multline*}
f(a=\sigma(z)) = \hat{y}
\end{multline*}
___________
<code>**Special cases:**</code>
 - In perceptron: activation function = threshold function
 - In linear regression: activation = net input = output

### Perceptron output
\begin{multline*}
\hat{y} = 
\begin{Bmatrix}
 0, & z \leq \theta \\ 
 1, & z > \theta 
\end{Bmatrix}
\end{multline*}

<code>**More convenient to re-arrange:**</code>
\begin{multline*}
\hat{y} = 
\begin{Bmatrix}
 0, & z -\theta \leq 0 \\ 
 1, & z -\theta > 0
\end{Bmatrix}
\end{multline*}

### General notation for signgle-layer neural networks
 - Common notation (moder texts): define the bias unit seperately
 - However oftern incovenient for mathematical notation
 
<code>**Seperate bias unit**</code>

\begin{multline*}
\sigma(\sum^{m}_{i=1} x_{i}w_{i} + b) = \sigma(\mathbf{x^{\top}}\mathbf{w}+b) = \hat{y}
\end{multline*}
 
\begin{multline*}
\sigma(z)=
\begin{Bmatrix}
 0, & z \leq \theta \\ 
 1, & z > \theta 
\end{Bmatrix}
\end{multline*}


\begin{multline*}
b = - \theta
\end{multline*}


Assume binary classification task, perceptron finds decision boundary if classes are separable. How acctualy it learns the weights to succeed in this task. This is problem with 2 features for simplicity. If we would have more features that would way more challenging to visualize. 

It starts at the random place and then basically update the rule on each iteration. At each iteration it will try to adjust the decision boundary to make the better. The circle indicates which shuffled data point currently it looks at a time. It it finds that there is a missclasification then it moves the desicion boundary. 

 - <code>**If correct:**</code>Do nothing if the prediction is equal to the target
 - <code>**If incorrect, scenario a):**</code>: If output is 0 and target is 1, `add` input vector to the weight vector
 - <code>**If incorrect, scenario b):**</code>: If output is 1 and target is 0, `subtract` input vector from weight vector

Guaranteed to converge (evnetually) if a soliution exists. If there is not soliution like if we would had a blue dot ammong orange dots , there would be no decision boundary where it could classify this data set perfectly. It would move forth and back the decision boundary always flipping.

<code>**compact mathematical meta algorithm expression of this rule**</code>

\begin{multline*}
\mathbf{D} = (\langle \mathbf{x}^{[1]}, y^{[1]} \rangle), \langle \mathbf{x}^{[2]}, y^{[2]} \rangle), \langle \mathbf{x}^{[n]}, y^{[n]} \rangle) \in (\mathbb{R}^{m} \times \{0,1\})^{n}
\end{multline*}

 - <code>**1.**</code> Intialize w:= 0^m <code>**(assume notation where wight includes bias)**</code>
 - <code>**2.**</code> For every training epoch:
   - <code>**A.**</code> For every \begin{multline*}\langle x^{[i]}, y^{[i]}\rangle \in \mathbf{D} :\end{multline*}
       - \begin{multline*} \hat{y} := \sigma(\mathbf{x^{[i]}}^{\top} \mathbf{w}) \end{multline*}
       - \begin{multline*} \text{err} := (y^{[i]} - \hat{y}^{[i]}) \end{multline*}
       - \begin{multline*} \mathbf{w} := \mathbf{w}  + \text{err} \times \mathbf{x}^{[i]} \end{multline*}

Epoch means pass over (iteration) the entire training set. If we have a training set like above then the epoch would be processing every data point in this set. <code>**Starting from first and until we reached the last data point then this 1 epoch**</code>. Then we may go back to the begining and do another sweep of data set and that would be the 2 epoch.
<br>
Then there is a inner loop for every data point in the data set we perfomr steps a, b, c. If we have multiple epochs we can do this multiple times. <code>**Perceptron processes 1 training example at a time.**</code> a) It computes prediction as a net input sum. 2) Compute error , its compted by true label minus predicted label. Assume simple case where we don't have any error y^[i] - y_hat^[i]  being 0's then the error is 0. c) Is weight update taking old weight pluse the error times the input vector. But the error is 0 then we cancel the input vector values and there is no update. 
<br>

\begin{multline*}
\mathbf{w} := \mathbf{w}  + \text{err}\hspace-2.2ex\diagup * x^{[i]} \hspace-2.2ex\diagup
\end{multline*}

And if both (target label and prediction) is 1 we can also cancel the weight update. So the weight update can happend only when there is an error. Let consider a case when we cave target label as 1 and prediction as 0 then the error is one. Then we are <code>**adding the input vector**</code>. And if reversed the target label 0 and our prediction is 1 then we have an error = -1. In this case we are <code>**subtracting the input vector**</code>

```python
def train_test_split(X, y):
    # Shuffling & train/test split
    shuffle_idx = np.arange(y.shape[0])
    shuffle_rng = np.random.RandomState(123)
    shuffle_rng.shuffle(shuffle_idx)
    X =  X.iloc[shuffle_idx]
    y =  y.iloc[shuffle_idx]

    X_train, X_test = X.iloc[:70], X.iloc[70:]
    y_train, y_test = y.iloc[:70], y.iloc[70:]
    
    return X_train, X_test, y_train, y_test
```

<code>**Defining perceptron model**</code>

We will implement perceptron using <code>**forward**</code> and <code>**backward**</code> methods. 

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

class Perceptron():
    def __init__(self, num_features):
        self.num_features = num_features
        self.weights = torch.zeros( num_features, 1, dtype=torch.float32, device=device)
        self.bias = torch.zeros(1, dtype=torch.float32, device=device)
        
        self.ones = torch.ones(1)
        self.zeros = torch.zeros(1)

    def forward(self, x):
        # computing net input
        linear = torch.mm(x, self.weights) + self.bias
        # threshold function that outputs either 1 or 0 
        predictions = torch.where(linear > 0., self.ones, self.zeros)
        return predictions
        
    def backward(self, x, y):  
        # computing the error by using pseudo backpropagation 
        predictions = self.forward(x)
        errors = y - predictions
        return errors
        
    def train(self, x, y, epochs):
        for e in range(epochs):
            for i in range(y.shape[0]):
                errors = self.backward(x[i].reshape(1, self.num_features), y[i]).reshape(-1)
                self.weights += (errors * x[i]).reshape(self.num_features, 1)
                self.bias    += errors
                
    def evaluate(self, x, y):
        predictions = self.forward(x).reshape(-1)
        accuracy = torch.sum(predictions == y).float() / y.shape[0]
        return accuracy
```

<code>**Training the Perceptron**</code>

```python
ppn = Perceptron(num_features=2)

X_train_tensor = torch.tensor(X_train_N, dtype=torch.float32, device=device)
y_train_tensor = torch.tensor(y_train,   dtype=torch.float32, device=device)

ppn.train(X_train_tensor, y_train_tensor, epochs=5)

print('Model parameters:')
print('  Weights: %s' % ppn.weights)
print('  Bias: %s' % ppn.bias)
```

<code>**Evaluating the model**</code>

```python
X_test_tensor = torch.tensor(X_test_N, dtype=torch.float32, device=device)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32, device=device)

test_acc = ppn.evaluate(X_test_tensor, y_test_tensor)
print('Test set accuracy: %.2f%%' % (test_acc*100))
```

##  4.2: Perceptron convergence theorem

## 5: The geometric intuition behind the prerceptron

Recal the perceptron learning algoritm when there is no error it does no update and if incorrect there two scenarios that lead to either addition or substraction of input vector from weights. 

<code>**Why weight vector is perpendicular to the decision boundary ?**</code>

\begin{multline*}
\left\{\begin{matrix}
 0, &  w^{\top} x \leq 0 \\ 
 1, &  w^{\top} x > 0
\end{matrix}\right.
\end{multline*}

<br>

If we write the dot product a little bit differntly as the <code>**magnitudes**</code> of these two vectors and then the angle between them cos() then we have the following equation that you might have seen in linear algebra:

\begin{multline*}
 w^{\top} x  = \| w \| \cdot \|x\| \cdot \cos(\theta)
\end{multline*}

<br>

<code>**cos()**</code> needs to be 0 at the boundary and it is zero at 90

So if we want this dot product to be 0 we have to have at least one of those two term as 0 but this is not the case (vector is 0 if its all valueas are 0). Another case is when the angle between weight vector and decision boundary , - cos() is 0. And this term is 0 when we have 90 between the two, cos(90) = 0.

<code>**What else does this mean ?**</code>

Thinking about this 90 angle we can deduce that there will be an angle smaller then 90 on this blue side of the decision boundary. If we have a data point (feature vector) that is somewhere on right side of the decision boundary  it will have an angle with the weight vector that is < 90. And it doesn't matter how large is the weight/input vector. If we would scale it by 1000 it will still get the same classification results and maintai the angles (but large inputs will take much longer to converge if you check the bounds we defined previously). But the learning would be much faster if the data points are scaled that why we use <code>**standartization**</code> methods in code. So the take away is that the angle is smaller than 90

Focus on the actual prediction case we have <code>**correct and wrong sides**</code> assume that we have an input vector with class label 1. And weight vector for the correct prediction must be somehwere such that the angle is smaller then 90 to make correct prediction.

The dot product will then be positive , i.e > 0 since,

\begin{multline*}
 w^{\top} x  = \| w \| \cdot \|x\| \cdot \cos(\theta)
\end{multline*}

As well we have a red vector that is not ideal where we make a wrong prediction because the angle is not smaller than 90 it is larger. If it is between larger then 90  and - 270 we will make wrong prediction , the value of cos() would be negative. We would be smaller than 0 and make prediction for 0.
<br>
How to correct this prediction, it is by moving the weight vector closer to the side of training data point. In perceptron algoritm we add feature vector to the weight vector.
